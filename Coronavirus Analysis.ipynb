```python
## Now lets work with Time Series Data
```

```python
#we have three sets to work with
# df_time_confirmed
# df_time_deaths
# df_time_recovered
df_time_confirmed.head()
```

```python
#need to divide up into individual observations
#start by removing province
df_time_confirmed = df_time_confirmed.drop(columns=['Province/State'])
df_time_confirmed.head()
```

```python
df_time_deaths = df_time_deaths.drop(columns=['Province/State'])
df_time_recovered = df_time_recovered.drop(columns=['Province/State'])

df_deaths = df_time_deaths.melt(id_vars=['Country/Region','Lat','Long'],var_name='Date',value_name='Deaths')
df_recovered = df_time_recovered.melt(id_vars=['Country/Region','Lat','Long'],var_name='Date',value_name='Recovered')


df_conf = df_time_confirmed.melt(id_vars=['Country/Region','Lat','Long'],var_name='Date',value_name='Confirmed')
df_conf.head()
```

```python
df_conf.tail()
```

```python
df_conf = df_conf.drop(columns=['Lat','Long'])
#df_conf = df_conf.groupby(['Country/Region','Date'])[['Confirmed']].sum()
df_deaths = df_deaths.groupby(['Country/Region','Date'], as_index=False)['Deaths'].sum()
df_recovered = df_recovered.groupby(['Country/Region','Date'], as_index=False)['Recovered'].sum()


df_conf = df_conf.groupby(['Country/Region','Date'], as_index=False)['Confirmed'].sum()
df_conf.head()
```

```python
df_conf['UniqueID'] = df_conf['Country/Region']+df_conf['Date'].astype(str)
df_deaths['UniqueID'] = df_deaths['Country/Region']+df_deaths['Date'].astype(str)
df_recovered['UniqueID'] = df_recovered['Country/Region']+df_recovered['Date'].astype(str)

df_step = df_conf.merge(df_deaths, on='UniqueID')
df_combo = df_step.merge(df_recovered,on='UniqueID')
df_combo = df_combo.drop(columns=['Country/Region_x','Date_x','Country/Region_y','Date_y','UniqueID'])
df_combo.head()
```
